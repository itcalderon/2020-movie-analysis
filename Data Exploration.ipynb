# Import the Box Office Mojo Data Set

```python
df = pd.read_csv('Data/bom.movie_gross.csv')
print(df.shape)
df.head()

```

```python
df.describe()
```

```python
df.info()
```

```python
# Curious to see which movies top the list as most lucrative, domestically.
df.sort_values(by='domestic_gross', ascending=False).head()
```

```python
# Which studio makes the most revenue?
df.groupby('studio').domestic_gross.sum().sort_values(ascending=False).head(5)
```

```python
# Which Studio makes the most movies?
df.groupby('studio').title.count().sort_values(ascending=False)
```

```python
# Mean domestic gross revenue
df.groupby('studio').domestic_gross.mean().sort_values(ascending=False)
```

# Import Rotten Tomatoes data set.


```python
rt_movie = pd.read_csv('Data/rt.movie_info.tsv', sep='\t')
print(rt_movie.shape)
rt_movie.head()
```

```python
rt_movie.id.nunique()
```

```python
rt_reviews = pd.read_csv('Data/rt.reviews.tsv', sep='\t', encoding='latin1')
print(rt_reviews.shape)
rt_reviews.head(10)
```

```python
rt_reviews.id.nunique()
```

```python
rt_reviews.isna().sum()
```

```python
rt_merged = pd.merge(rt_reviews, rt_movie, on='id')
print(rt_merged.shape)
rt_merged.head()
```

```python
rt_merged.to_csv('Data/rotten_tomatoes_merged.csv')
```
