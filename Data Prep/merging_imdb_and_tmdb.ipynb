```python
import pandas as pd
```

```python
df1 = pd.read_csv('Cleaned Data/cleaned_imdb-basics-and-ratings.csv')
df1.drop(['Unnamed: 0'], axis=1, inplace=True)
df1.head()
```

```python
df2 = pd.read_csv('Cleaned Data/imdb.movies_basics_and_crew.csv')
df2.drop(['Unnamed: 0'], axis=1, inplace=True)
df2.head()
```

```python
df3 = pd.read_csv('Cleaned Data/cleaned_tmdb-num-bom.csv')
df3.drop(['Unnamed: 0'], axis=1, inplace=True)
df3.head()
```

```python
df4 = pd.merge(df1, df3, left_on='title_id', right_on='imdb_id', how='right')
df4.info()
```

```python
df4.head()
```

```python
df4.drop(['primary_title', 'original_title', 'genres', 'title_id'], axis=1, inplace = True)
```

```python
df4 = df4.rename(columns={'averagerating': 'imdb_rating', 'numvotes': 'imdb_numvotes'})
```

```python
df4.head(3)
```

```python
df4.to_csv('Cleaned Data/final_merged.csv')
```
